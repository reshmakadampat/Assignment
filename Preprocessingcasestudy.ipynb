```python
missing_val_col=['Sale Price','No of Bathrooms','Flat Area (in Sqft)','Lot Area (in Sqft)','Area of the House from Basement (in Sqft)',
                 'Zipcode','Latitude','Longitude','Living Area after Renovation (in Sqft)']
```

Handling outliers

```python
large_unique_col=['Sale Price','Flat Area (in Sqft)','Lot Area (in Sqft)','Area of the House from Basement (in Sqft)','Basement Area (in Sqft)',
                  'Age of House (in Years)','Latitude','Longitude','Living Area after Renovation (in Sqft)',
                  'Lot Area after Renovation (in Sqft)']
```

```python
for cols in large_unique_col:
    sns.boxplot(data=df[cols] ,orient= "h")
    plt.show()
```

```python
outliers_percentages = []

for col in large_unique_col:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers= ((df[col] < lower_bound) | (df[col] > upper_bound))
    df[col] = df[col].where(~outliers, np.median(df[col]))
```

Feature Scaling

```python
#splitting dataset into features and targets
x=df1.drop('Sale Price',axis=1)
y=df1['Sale Price']
```

```python
#applying min mxa scaling on numerical cols
x1=x.copy()
x1.columns
```

```python
#dropping the encoded cols from x1
x1.drop(['Waterfront View','Condition of the House'],axis=1,inplace=True)
```

```python
x1.columns
```

```python
#doing min max scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x1=scaler.fit_transform(x1)
```

```python
#we have to convert this array into dataframe
x1=pd.DataFrame(x1,columns=['No of Bedrooms', 'No of Bathrooms', 'Flat Area (in Sqft)',
       'Lot Area (in Sqft)', 'No of Floors', 'Overall Grade',
       'Area of the House from Basement (in Sqft)', 'Basement Area (in Sqft)',
       'Age of House (in Years)', 'Renovated Year', 'Zipcode', 'Latitude',
       'Longitude', 'Living Area after Renovation (in Sqft)',
       'Lot Area after Renovation (in Sqft)'])
x1.head()
```

```python
#merging this scaled data
x2=x.drop(['No of Bedrooms', 'No of Bathrooms', 'Flat Area (in Sqft)',
       'Lot Area (in Sqft)', 'No of Floors', 'Overall Grade',
       'Area of the House from Basement (in Sqft)', 'Basement Area (in Sqft)',
       'Age of House (in Years)', 'Renovated Year', 'Zipcode', 'Latitude',
       'Longitude', 'Living Area after Renovation (in Sqft)',
       'Lot Area after Renovation (in Sqft)'],axis=1)
x=pd.concat([x1,x2],axis=1)
x.head()
```

```python
plt.figure(figsize=(5,8))
sns.heatmap(df1.corr()['Sale Price'].sort_values(ascending=False).to_frame(),
            annot=True, vmin=-1, vmax=1, linewidths=0.3, linecolor='black', cmap= 'flare')
plt.title('Correlation of Variables to the Sale Price Column')
plt.show()
```

Since variables in our dataset do not have a strong correlation with the target variable, traditional linear models like linear regression might not perform well.

Linear regression model

```python
#splitting the dataset into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#creating a basic linear regression model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#Making prediction and checking the r squared value
y_pred=lr.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
#if the r squared value is close to 1,it means the model is good
```

XGBoost regression model

it is an optimized version of gradient boosting that is highly efficient and scalable, often used in machine learning competitions.

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
simple_reg= xgb.XGBRegressor(objective="reg:squarederror", seed= 123)
simple_reg.fit(x_train,y_train)

y_pred = simple_reg.predict(x_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(rmse)
```
